```python
import boto3
import json
from pathlib import Path
```

```python
s3 = boto3.resource('s3')

content_object = s3.Object('estates-9036941568', 'data/bronze/20210508-211200_2910227036.json')
file_content = content_object.get()['Body'].read().decode('utf-8')
json_content = json.loads(file_content)
```

```python
df = pd.DataFrame.from_dict(json_content, orient='index')
df.head()
```

```python
new_col_names = ['estate_id'] + df.columns.to_list()
```

```python
df = df.reset_index()
```
